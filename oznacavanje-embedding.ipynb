```python
import tensorflow as tf
```

```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
#recenice
recenice = [ "the glass of milk",
            "the glass of juice",
            "the cup of tea",
            "I am a good boy",
            "I am a good developer",
            "understand the meaning of words",
            "your videos are good" 
    ]
```

```python
#definisanje obima rijecnika
obim_rijeci = 10000
```

```python
#Jedna Vruca Reprezentacija - One Hot Representation
jedna_vruca_repr = [one_hot(rijeci,obim_rijeci) for rijeci in recenice]
jedna_vruca_repr
```

```python
#Reprezentacija Oznacavanja Rijeci - Word Embedding Representation
from tensorflow.keras.layers import Embedding
#from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
#Ovdje je princip da sve recenice postanu ravnopravne kako bi bile prikladne za treniranje u RNN model, tako Å¡to se odredi
#duzina za sve recenice, te se vrsi popuna recenica nulama, tako da postanu identicne duzine...
#parametar padding, sa argumentom "pre" stavlja nule na pocetak, a sa "post" na kraj recenica.
duzina_recenice = 8
oznaceni_dokumenti = pad_sequences(jedna_vruca_repr, padding="pre",maxlen=duzina_recenice)
oznaceni_dokumenti
```

```python
#Reprezentacija Osobina
dimenzija = 10
```

```python
#Stvaranje Modlea sa Slojem za oznacavanje
model = Sequential()

#dodavanje tog Embedding sloja modelu
model.add(Embedding(obim_rijeci,dimenzija,input_length=duzina_recenice))

#Sacinjavanje modela -"mse" jeste funkcija gubitka, dok "adam" je optimizer
model.compile("adam","mse")
```

```python
#PREGLED modela
model.summary()
```

```python
model.predict(oznaceni_dokumenti)
```
